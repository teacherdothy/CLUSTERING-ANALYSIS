```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("FOR_CLUSTERING.csv")
```

```python
df.head()
```

```python
plt.scatter(df['Pre-Review'],df['Comprehensive'],df['Post-Review'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['Pre-Review','Comprehensive','Post-Review']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1['Pre-Review'],df1['Comprehensive'],df1['Post-Review'],color='green')
plt.scatter(df2['Pre-Review'],df2['Comprehensive'],df2['Post-Review'],color='yellow')
plt.scatter(df3['Pre-Review'],df3['Comprehensive'],df3['Post-Review'],color='blue')

plt.xlabel('Pre-Review')
plt.ylabel('Post-Review')
plt.legend()
```

```python
scaler = MinMaxScaler()
scaler.fit(df.Pre-Review)
df.Pre-Review = scaler.transform(df.Pre-Review)

scaler.fit(df.Post-Review)
df.Post-Review = scaler.transform(df.Post-Review)
df
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1['Pre-Review'],df1['Comprehensive'],df1['Post-Review'],color='green')
plt.scatter(df2['Pre-Review'],df2['Comprehensive'],df2['Post-Review'],color='yellow')
plt.scatter(df3['Pre-Review'],df3['Comprehensive'],df3['Post-Review'],color='blue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],km.cluster_centers_[:,2],color='red',marker='*',label='centroid')
plt.legend()
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Pre-Review','Comprehensive','Post-Review']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse,marker = 'o')
```
